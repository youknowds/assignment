```python
import pandas as pd

df = pd.read_excel("C:/Users/kimso/Desktop/한시트에 2004-2020.xlsx")
df
```

```python
df_use=df.drop(['전국'])
df_use
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(rc={'figure.figsize':(10,8)})
plt.hist(df_use['선거구'],bins=5,color='salmon')
plt.xlabel("number of constituency")
plt.show()
```

```python
plt.figure(figsize=(10,10))
correlation_matrix=df_use.corr().round(3)
sns.heatmap(data=correlation_matrix,annot=True, cmap='bwr')
```

```python
sns.set_style('darkgrid')
fig,axs=plt.subplots(ncols=3)
sns.regplot(x='인구밀도비',y='선거구 비율',data=df_use,ax=axs[0])
sns.regplot(x='1인당 민간소비비',y='선거구 비율',data=df_use,ax=axs[1])
sns.regplot(x='인구 비',y='선거구 비율',data=df_use,ax=axs[2])
```
